```python
def calculate_bill(amount, gst_rate = 0.18, discount = 0):
    gst = amount * gst_rate # Bill with gst value
    total = amount + gst - discount

    print(f"Amount: Rs. {amount}")
    print(f"GST ({gst_rate * 100} %): rs. {gst}")

    if discount > 0:
        print(f"Discount : - {discount}")
    print(f"Ttoal: Rs. {total}")
    print("-" * 25)
```

```python
def calculate_total(marks):
    total = 0
    for mark in marks:
        total = total + mark
    return total

def calculate_average(marks):
    total = calculate_total(marks)    # Using another function's result!
    return total / len(marks)

def get_grade(average):
    if average >= 90:
        return "A 🌟"
    elif average >= 70:
        return "B"
    elif average >= 50:
        return "C"
    else:
        return "Fail ❌"
```

```python
# Chain them together
marks = [85, 92, 78, 95, 88]

avg = calculate_average(marks)
grade = get_grade(avg)

print(f"Average: {avg}")
print(f"Grade: {grade}")
```

```python
def calculate_average(marks):
    # Clauclate the average of list of marks.
    """Calculate the average of a list of marks.
    """
    total = 0
    for mark in marks:
        total = total + mark
    return total / len(marks)

def get_grade(average):
    """
    Determine grade based on average marks.
       if students marks are greater than 90 --> we are gonna asisgn grade a and retiurn that value.
       if students marks are greater than 70 --> we are gonna asisgn grade b and retiurn that value.
       if students marks are greater than 50 --> we are gonna asisgn grade c and retiurn that value.
       if students marks are less than 50 --> we are gonna asisgn grade fail and retiurn that value.
    """
    if average >= 90:
        return "A 🌟"
    elif average >= 70:
        return "B"
    elif average >= 50:
        return "C"
    else:
        return "Fail ❌"

def print_report(name, marks):
    """Generate and print a student report card."""
    avg = calculate_average(marks)
    grade = get_grade(avg)

    print(f"Student: {name}")
    print(f"Marks: {marks}")
    print(f"Average: {avg}")
    print(f"Grade: {grade}")
    if "Fail" in grade:
        print("⚠️ Needs improvement")
    else:
        print("✅ Passed!")
    print("-" * 30)

# Now use it for ANY student — just one line each!
print_report("Rahul", [85, 92, 78, 95, 88])
print_report("Priya", [45, 52, 38, 60, 55])
print_report("Amit", [92, 95, 88, 97, 91])
```
