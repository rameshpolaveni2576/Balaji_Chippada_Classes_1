```python
# Level 1: Direct keys
print(restaurant["name"])        # Biryani Blues
print(restaurant["rating"])      # 4.2

# Level 2: Nested object
print(restaurant["location"]["city"])       # Hyderabad
print(restaurant["location"]["pincode"])    # 500034

# Level 2: Array (list) inside object
print(restaurant["menu"][0])    # {'item': 'Chicken Biryani', 'price': 350, ...}

# Level 3: Key inside array item
print(restaurant["menu"][0]["item"])     # Chicken Biryani
print(restaurant["menu"][0]["price"])    # 350
print(restaurant["menu"][1]["item"])     # Paneer Biryani

# Level 2: Nested object → array
print(restaurant["timings"]["days"])         # ['Mon', 'Tue', ...]
print(restaurant["timings"]["days"][0])
```

```python

# A more complex nested structure
classroom = {
    "class": "B.Tech Year 1",
    "semester": 2,
    "students": [
        {
            "name": "Rahul Sharma",
            "roll": 42,
            "marks": {
                "maths": 85,
                "physics": 92,
                "chemistry": 78
            },
            "attendance": 0.88
        },
        {
            "name": "Priya Patel",
            "roll": 15,
            "marks": {
                "maths": 90,
                "physics": 88,
                "chemistry": 95
            },
            "attendance": 0.95
        },
        {
            "name": "Amit Kumar",
            "roll": 7,
            "marks": {
                "maths": 72,
                "physics": 65,
                "chemistry": 80
            },
            "attendance": 0.72
        }
    ]
}

# Navigate and extract
print(f"📚 {classroom['class']} — Semester {classroom['semester']}")
print(f"Total students: {len(classroom['students'])}\n")
```

```python
import json

for student in classroom["students"]:
    marks = student["marks"]
    total = sum(marks.values())
    avg = total / len(marks)

    print(f"  {student['name']} (Roll #{student['roll']})")
    print(f"    Marks: M={marks['maths']}, P={marks['physics']}, C={marks['chemistry']}")
    print(f"    Average: {avg:.1f} | Attendance: {student['attendance']*100:.0f}%")
    print()

# Save the whole thing to a file
with open("classroom.json", "w") as file:
    json.dump(classroom, file, indent=4)
print("✅ Saved to classroom.json")
```

```python

# A more complex nested structure
classroom = {
    "class": "B.Tech Year 1",
    "semester": 2,
    "students": [
        {
            "name": "Rahul Sharma",
            "roll": 42,
            "marks": {
                "maths": 85,
                "physics": 92,
                "chemistry": 78
            },
            "attendance": 0.88
        },
        {
            "name": "Priya Patel",
            "roll": 15,
            "marks": {
                "maths": 90,
                "physics": 88,
                "chemistry": 95
            },
            "attendance": 0.95
        },
        {
            "name": "Amit Kumar",
            "roll": 7,
            "marks": {
                "maths": 72,
                "physics": 65,
                "chemistry": 80
            },
            "attendance": 0.72
        }
    ]
}


```

```python
def add_grace_marks(marks):
    """Step 1 (innermost): Add 5 grace marks"""
    return marks + 5

def calculate_percentage(marks):
    """Step 2: Convert marks to percentage (out of 200)"""
    graced = add_grace_marks(marks)
    return (graced / 200) * 100

def assign_grade(marks):
    """Step 3: Give a grade based on percentage"""
    percentage = calculate_percentage(marks)
    if percentage >= 90:
        return percentage, "A+"
    elif percentage >= 75:
        return percentage, "A"
    elif percentage >= 50:
        return percentage, "B"
    else:
        return percentage, "F"

def print_result(name, marks):
    """Step 4 (outermost): Print the full result"""
    percentage, grade = assign_grade(marks)
    print(f"{name}: {marks} marks → {percentage:.1f}% → Grade {grade}")
```

```python
print_result("Rahul", 150)
```

🧩 Real-life analogy

Imagine a job screening system:

A candidate must:

Have age > 18
Have experience > 2 years
Have degree

Each of these is a rule (function).

Your engine:

Stores these rules
Runs all rules on a candidate
If ALL pass → ✅ accept
If ANY fails → ❌ reject
🏗️ What you need to build
1. Main function
create_rule_engine()

This is like a factory that creates your rule system.

2. Inside this function
👉 A list to store rules
This list should live inside the function
This is your closure state

Example idea:

rules = [rule1, rule2, rule3]
3. Inner function: add_rule(rule_func)

Purpose:
👉 Add a new rule to the system

Input:

A function (rule_func)

What it does:

Stores this function in your rules list
4. Inner function: evaluate(value)

Purpose:
👉 Check if value satisfies all rules

What it does:

Loop through all stored rules
Apply each rule to the value
If any rule returns False → stop and return False
If all return True → return True
🔁 Flow of usage (IMPORTANT)

This is how someone will use your system:

engine = create_rule_engine()

engine.add_rule(rule1)
engine.add_rule(rule2)
engine.add_rule(rule3)

result = engine.evaluate(some_value)
🧪 Example rules (concept only)

You might define rules like:

is_even(x)
greater_than_10(x)
less_than_100(x)

Then:

engine.evaluate(20) → True
engine.evaluate(5) → False

```python
rules = []

def is_even(x):
    return x % 2 == 0

def greater_than_10(x):
    return x>10

def less_than_100(x):
    return x<100

rules.append(is_even)
rules.append(greater_than_10)
rules.append(less_than_100)
```

```python
def create_rule_engine():
    rules = []

    def add_rule(balaji):
        rules.append(balaji)
        print(f"Rules added {balaji}")

    def evaluate(value):
        for rule in rules:
            if not rule(value):
                return False
        return True
    print(rules)
    return add_rule, evaluate

add_rule, evaluate = create_rule_engine()

def is_even(x):
    return x % 2 == 0

def greater_than_10(x):
    return x>10

def less_than_100(x):
    return x<100


add_rule(is_even)
add_rule(greater_than_10)
add_rule(less_than_100)


print(evaluate(99))
```
